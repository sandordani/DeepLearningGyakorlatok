```python
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt
import numpy as np
```

```python
#task 1
images=[]
for i in range(5):
    images.append(mpimg.imread("images/im%d.jpg" % (i+1)))
```

```python
reds = []
greens = []
blues = []
for i in images:
    red = np.array(i.copy()[:,:,0],dtype = "float")
    green = np.array(i.copy()[:,:,1],dtype = "float")
    blue = np.array(i.copy()[:,:,2],dtype = "float")
    print("red:")
    print(red)
    plt.imshow(red)
    plt.show()
    print("green:")
    print(green)
    plt.imshow(green)
    plt.show()
    print("blue:")
    print(blue)
    plt.imshow(blue)
    plt.show()
    reds.append(red)
    greens.append(green)
    blues.append(blue)
```

```python
#task 2
mean = np.zeros((3,128,128)) #mean for channels red (0), green (1) and blue (2)
for i in range(5):
    mean[0] += reds[i]
    mean[1] += greens[i]
    mean[2] += blues[i]
mean /= 5
print(mean)
```

```python
deviation = np.zeros((3,128,128)) #deviance for channels red (0), green (1) and blue (2)
for i in range(5):
    deviation[0] += (mean[0] - reds[i]) ** 2
    deviation[1] += (mean[1] - greens[i]) ** 2
    deviation[2] += (mean[2] - blues[i]) ** 2
deviation = np.sqrt(deviation/5)
print(deviation)
```

```python
for i in range(5):
    reds[i] -= mean[0]
    greens[i] -= mean[1]
    blues[i] -= mean[2]
for i in range(5):
    reds[i] /= deviation[0]
    greens[i] /= deviation[1]
    blues[i] /= deviation[2]
```

```python
newMean = np.zeros((3,128,128)) #new mean for channels red (0), green (1) and blue (2) (should be 0)
for i in range(5):
    newMean[0] += reds[i]
    newMean[1] += greens[i]
    newMean[2] += blues[i]
newMean /= 5
if(np.array_equiv(np.round(newMean,5), np.zeros((3,128,128)))):
    print("Mean adjusted to 0")
else:
    print("Failed to adjust mean")
    print(newMean)
    
newDeviation = np.zeros((3,128,128)) #new deviance for channels red (0), green (1) and blue (2) (should be 0)
for i in range(5):
    newDeviation[0] += (newMean[0] - reds[i]) ** 2
    newDeviation[1] += (newMean[1] - greens[i]) ** 2
    newDeviation[2] += (newMean[2] - blues[i]) ** 2
newDeviation = np.sqrt(newDeviation/5)

if(np.array_equiv(np.round(newDeviation,5), np.ones((3,128,128)))):
    print("Deviation adjusted to 1")
else:
    print("Failed to adjust deviation")
    print(newDeviation)
```

```python
#task 3
from bs4 import BeautifulSoup
from urllib import request
import string
```

```python
url = "https://blog.keras.io/the-future-of-deep-learning.html"
r = request.urlopen(url)
webContent = r.read()
soup = BeautifulSoup(webContent)
print(soup.text)
```

```python
letters = {}
for letter in soup.text:
    if(letter.lower() in string.ascii_lowercase):
        if(letter.lower() not in letters):
            letters[letter.lower()] = 1
        else:
            letters[letter.lower()] += 1
print(letters)
```

```python
plt.bar(letters.keys(), letters.values())
plt.show()
```

```python
#task 4
from scipy import signal
from scipy.io import wavfile
```

```python
sample_rate, samples = wavfile.read('sound.wav')
frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate)
```

```python
plt.pcolormesh(times, frequencies, spectrogram)
plt.imshow(spectrogram)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
